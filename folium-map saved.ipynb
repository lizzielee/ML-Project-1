```python
locationlist = water1[["lat","lon"]].values.tolist()
IDs = water1["Site_ID"].values.tolist()
labels = water1["labels"].values.tolist()
location = water1['lat'].mean(), water1['lon'].mean()
```

```python

water_dated_1 = water1.loc[water1['Date_Collected'] == '5/20/2016']
water_dated_2 = water1.loc[water1['Date_Collected'] == '9/11/2015']
water_dated_3 = water1.loc[water1['Date_Collected'] == '9/19/2014']
water_dated_4 = water1.loc[water1['Date_Collected'] == '9/13/2013']
```

```python
locationlist1 = water_dated_1[["lat","lon"]].values.tolist()
IDs = water_dated_1["Site_ID"].values.tolist()
labels = water_dated_1["labels"].values.tolist()
location = water_dated_1['lat'].mean(), water_dated_1['lon'].mean()

m1 = folium.Map(location=location, zoom_start=9)
for point in range(len(locationlist1)):
    popup = folium.Popup(IDs[point]+", "+str(labels[point]), parse_html=True)
    folium.CircleMarker(locationlist[point], popup=popup,color=colorMap[int(labels[point])]).add_to(m1)

#display(m1)
m1.save("5_20_2016.html")
```
