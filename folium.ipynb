```python
locationlist = water[["lat","lon"]].values.tolist()
IDs = water["Site_ID"].values.tolist()
labels = water["Discription"].values.tolist()
location = water['lat'].mean(), water['lon'].mean()

m = folium.Map(location=location, zoom_start=9)
for point in range(len(locationlist)):
    popup = folium.Popup(IDs[point]+", "+labels[point], parse_html=True)
    folium.Marker(locationlist[point], popup=popup).add_to(m)


display(m)
```
